```python
import numpy as np
import pandas as pd
#imported dataset from the kaggle
a=pd.read_csv("/content/swiggy.csv")
print(a.head())
```

# ***1.Number of restaurants within a city***

```python

top_10_cities=a['City']
city_counts=top_10_cities.value_counts().to_frame().reset_index()
print(city_counts)

#Conclusion:To find the number of restaurants in an city individually across india.
```

```python

import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(12,8))
ax.bar(city_counts['City'],city_counts['count'],col="Cyan")
ax.set_title("Number of restaurents within a city")
ax.set_xlabel("City")
ax.set_ylabel("Number of Restaurents")
plt.show()
```

# ***1 A).No of Restaurents within a area***

```python
top10_Areas=a['Area']
Area_Counts=top10_Areas.value_counts().to_frame().head(10).reset_index()
Area_Counts
#Conclusion:To find the number of restaurants within a area
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(20,8))
ax.bar(Area_Counts['Area'],Area_Counts['count'])
ax.set_xlabel("Areas ")
ax.set_ylabel("Number of Restaurents")
ax.set_title("Top 10 Areas with maximum number of restaurents")
```

# ***2.Correlation between price and ratings***

```python
import scipy .stats
Price=a['Price'].to_frame()
print(Price)
print(Price.values)
Price_Values=Price.values.reshape(-1)
print("Price Values are",Price_Values)
Ratings=a['Avg ratings'].to_frame()
print(Ratings)
Rating_Values=Ratings.values.reshape(-1)
print("Rating Values are",Rating_Values)
# Finding Correlation using Scipy:
print("Correlation between Price and Ratings")
scipy.stats.pearsonr(Price_Values,Rating_Values)

```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.scatter(a['Price'],a["Avg ratings"])
ax.set_title("Correlation between Price vs Ratings")
ax.set_xlabel("Prices")
ax.set_ylabel("Avg Ratings")
# Conclusion: The value of correlation is Positive (i.e) 0.113(Weak Correlation).Here a positive coefficient
#indicates a positive correlation.As price increases ratings tends to increase
```

# ***3.Correlation between Ratings vs Delivery Time***

```python
Delivery_Time=a['Delivery time'].to_frame()
print(Delivery_Time)
Ratings=a['Avg ratings'].to_frame()
print(Ratings)
# Finding Correlation:
Delivery_Time_Values=Delivery_Time.values.reshape(-1)
print("Delivery time values are",Delivery_Time_Values)
Rating_Values=Ratings.values.reshape(-1)
print("Rating Values are",Rating_Values)
print("Correlation between Delivery Time vs Ratings")
scipy.stats.pearsonr(Delivery_Time_Values,Rating_Values)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.scatter(a["Delivery time"],a["Avg ratings"])
ax.set_title("Correlation between Delivery time vs Ratings")
ax.set_xlabel("Delivery time")
ax.set_ylabel("Avg Ratings")
#Conclusions:The value of correlation is negative (i.e) -0.1469.Here a negative coefficient
#indicates a negative correlation.As Delivery time increases ratings tends to decrease.
```

# ***6.Normal Distribution of the ratings of restaurents in a city***

```python
import statistics
import numpy as np
Average_Ratings=a['Avg ratings']
print(Average_Ratings.shape)
Mean=statistics.mean(Average_Ratings)
print("Mean is =",Mean)
Standard_Deviation=statistics.stdev(Average_Ratings)
print("Standard Deviation is =",Standard_Deviation)
x=np.random.normal(Mean,Standard_Deviation,8680)
plt.hist(x,100,color="skyblue")
plt.title("Normal Distribution of the ratings of restaurents in a city")
plt.show()
#Conclusion: On an average the ratings of restaurants were lies between 3-4.
```
